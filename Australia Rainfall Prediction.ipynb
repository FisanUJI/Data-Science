# Feature Scaling

##### The final step of the machine learning model is feature scaling. It is a method to Standardize the training dataset in a specific range. In feature scaling, all the values are kept in the same range and on the same scale so that no variable dominates the other variable. 

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X = sc.fit_transform(X)
```

```python
# Visualize all the independent variables (X) as a dataframe
columns = ['Location', 'MinTemp', 'MaxTemp', 'Rainfall', 'WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Temp9am', 'Temp3pm', 'RainToday']
X = pd.DataFrame(data=X, columns= columns, dtype=None, copy=False)
X.head(10)
```

# Splitting Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(X_train)
```

```python
print(Y_train)
```

# Training the Model

> Importing the measurement metrics (e.g accuracy score, confusion matrix and classification report)

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

> By definition a confusion matrix C is such that C_{i, j}
is equal to the number of observations known to be in group i and predicted to be in group j.

> Thus in binary classification, the count of true negatives is
C_{0,0}, false negatives is C_{1,0}, true positives is
C_{1,1} and false positives is C_{0,1}.

### 1. Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, Y_train)
y_predict= lr.predict(X_test)

# Confusion Matrix for the Logistic Regression
print('Confusion Matrix (Logistic Regression):')
print(confusion_matrix(Y_test,y_predict))
print("\n*************************************************************\n")

# Classification Report for the Logistic Regression
print('Classification Report (Logistic Regression):\n' + classification_report(Y_test,y_predict))
print("*************************************************************\n")

# Accuracy Score for the Logistic Regression
print('Accuracy Score (Logistc Regression):\n')
print(accuracy_score(Y_test,y_predict))
```

### 2. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
y_predict= dt.predict(X_test)

# Confusion Matrix for the Decision Tree Classifier
print('Confusion Matrix (Decision Tree Classifier):')
print(confusion_matrix(Y_test,y_predict))
print("\n*************************************************************\n")

# Classification Report for the Decision Tree
print('Classification Report (Decision Tree):\n' + classification_report(Y_test,y_predict))
print("*************************************************************\n")

# Accuracy Score for the Decision Tree
print('Accuracy Score (Decision Tree):\n')
print(accuracy_score(Y_test,y_predict))
```

### 3. Random Forest Classifier





```python
from sklearn.ensemble import RandomForestClassifier
cl = RandomForestClassifier(n_estimators =150, random_state=0)
cl.fit(X_train,Y_train)
y_predict= cl.predict(X_test)

# Confusion Matrix for Random Forest
print('Confusion Matrix (Random Forest):')
print(confusion_matrix(Y_test,y_predict))
print("\n*************************************************************\n")

# Classification Report for the Random Forest
print('Classification Report (Random Forest):\n' + classification_report(Y_test,y_predict))
print("*************************************************************\n")

# Accuracy Score for the Random Forest
print('Accuracy Score (Random Forest):\n')
print(accuracy_score(Y_test,y_predict))
```

> From our confusion matrix above, there are 21697 True positive (TP) values and 1029 False Positive (FP) values

# 4. XGBoost Classifier

```python
import xgboost as xgb
xgb = xgb.XGBClassifier()
xgb.fit(X_train,Y_train)
y_predict= xgb.predict(X_test)

# Confusion Matrix for XGBoost Classifier
print('Confusion Matrix (XGBoost Classifier):')
print(confusion_matrix(Y_test,y_predict))
print("\n*************************************************************\n")

# Classification Report for the XGBoost Classifier
print('Classification Report (XGBoost Classifier):\n' + classification_report(Y_test,y_predict))
print("*************************************************************\n")

# Accuracy Score for the XGBoost Classifier
print('Accuracy Score (XGBoost Classifier):\n')
print(accuracy_score(Y_test,y_predict))
```
